```python
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
```

```python
tips = sns.load_dataset('tips')
```

```python
flights.head()
```

```python
tips.corr(numeric_only=True)
```

```python
sns.heatmap(tips.corr(numeric_only=True),cmap='coolwarm',annot=True)
```

```python
fp = flights.pivot_table(index='month',columns='year')
```

```python
sns.heatmap(fp)
```

```python
sns.clustermap(fp,figsize=(5,5))
```

```python
pg = sns.PairGrid(tips)
pg.map_diag(sns.kdeplot)
pg.map_lower(sns.scatterplot)
pg.map_upper(sns.violinplot)
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g.map(sns.scatterplot,'total_bill','day')
```
